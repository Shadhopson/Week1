# Looking into whether or not there are trends in when people come in for open hours

```python
feb= pd.read_csv("Time_of_day_data/Welcome Table Sign in 2020 - FEB.csv")
```

```python
jan = pd.read_csv("Time_of_day_data/Welcome Table Sign in 2020 - JAN.csv")
```

```python
db = pd.concat([jan,feb], axis=0)
```

```python
db.dropna(subset=["Date/Time"], inplace=True)
```

```python
db['Date/Time'] = pd.to_datetime(db['Date/Time'])
```

```python
db['Hour'] = db['Date/Time'].dt.hour
```

```python
db['Weekday'] = db['Date/Time'].dt.weekday +1
```

```python
c = db.groupby(['Weekday', 'Hour']).count()
```

```python
c
```

```python
s = db.groupby(['Weekday', 'Hour']).sum()
```

```python
s
```

```python
s['# in Group']/c['# in Group']
```

```python
(s['# in Group']/c['# in Group']).mean()
```
