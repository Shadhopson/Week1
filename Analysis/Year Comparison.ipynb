```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
summed_by_date = summed_by_date.rename(columns={'Unnamed: 0':'date'})
```

```python
summed_by_date['date'] =  pd.to_datetime(summed_by_date['date'])
```

```python
summed_date_graph = summed_by_date[['date', 'total_visitors']]
```

```python
summed_date_graph.index = summed_date_graph['date']
```

```python
del summed_date_graph['date']
```

```python
pv2 = pd.pivot_table(summed_date_graph, index=summed_date_graph.index.month, columns=summed_date_graph.index.year,
                    values='total_visitors', aggfunc='sum')
```

```python
pv2 = pv2.drop(2108,axis=1)
```

```python
week_graph = pv2 = pd.pivot_table(summed_date_graph, index=summed_date_graph.index.week, columns=summed_date_graph.index.year,
                    values='total_visitors', aggfunc='sum')
```

```python
week_graph = week_graph.drop(2108, axis=1)
```

```python
week_graph.plot(title="Year Comparison: Visitors Per Week")
plt.xlabel("Week")
plt.ylabel("Number of Visitors")
```
