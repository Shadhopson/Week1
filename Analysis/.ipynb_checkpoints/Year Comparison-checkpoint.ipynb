```python
import pandas as pd
summed_by_date = pd.read_csv('../Data_And_Data_Creation/advanced_welcome_table_summed.csv')
```

```python
print "   Average Per Day Values"
by_year_data = summed_by_date.groupby('year', as_index=False).mean()
by_year_data.to_csv('../Figures/Average_Per_Day_By_Year.csv')
by_year_data
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
summed_by_date = summed_by_date.rename(columns={'Unnamed: 0':'date'})
```

```python
summed_by_date['date'] =  pd.to_datetime(summed_by_date['date'])
```

```python
summed_date_graph = summed_by_date[['date', 'total_visitors']]
```

```python
summed_date_graph.index = summed_date_graph['date']
```

```python
del summed_date_graph['date']
```

```python
pv2 = pd.pivot_table(summed_date_graph, index=summed_date_graph.index.month, columns=summed_date_graph.index.year,
                    values='total_visitors', aggfunc='sum')
```

```python
pv2.plot(title="Year Comparison: Visitors Per Month")
plt.xlabel("Month")
plt.ylabel("Number of Visitors")

```

```python
summed_by_date.columns
```

```python
weekend = summed_by_date[summed_by_date['day'].isin([5,6])]
weekend.index = weekend['date']
```

```python
pv3 = pd.pivot_table(weekend, index=weekend.index.month, columns=weekend.index.year,
                    values='total_visitors', aggfunc='sum')
```

# Weekend Comparison

```python
pv3.plot()
plt.title("Year Comparison: Visits On Weekends")
plt.xlabel("Month")
plt.ylabel("Number of Visitors")
#plt.legend(loc="upper right")
```

```python
weekdays = summed_by_date[summed_by_date['day'].isin([1,3])]
weekdays.index = weekdays['date']
```

```python
pv4 = pd.pivot_table(weekdays, index=weekdays.index.month, columns=weekdays.index.year,
                    values='total_visitors', aggfunc='sum')
```
