```python
the_date = datetime.datetime.today()
weather_by_day = [{}]*24
for i in range(800):
    date_str =  str(the_date.date())
    the_date = the_date - datetime.timedelta(days=1)
    url = "https://api.darksky.net/forecast/6ee0b0521cccd43b16a44a63eba7cd8d/40.6987,-73.9859,"+ date_str+"T15:00:00"
    weather_data  = json.load(urllib.urlopen(url))
    try:
        for i in range(0,24):
            weather_by_day[i][date_str] =  weather_data['hourly']['data'][i]
    except:
        pass

```

```python
print weather_by_day
```

```python
for hour in range(0,24):
    with open('hourly_weather_data/hourly_past_weather'+str(hour)+'.json', 'a') as f:
        json.dump(weather_by_day[i], f)
```
